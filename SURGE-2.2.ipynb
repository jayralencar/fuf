# Defining an Input Functional Description

The easiest way to define and test the realization of an input Functional description (FD) is to use the def-test macro:

```common-lisp
(def-test t1
  "This car is expensive."
  ((cat clause)
   (proc ((type ascriptive)
      (mode attributive)))
   (partic ((carrier ((lex "car")
                      (cat common)
                      (distance near)))
        (attribute ((lex "expensive")
                    (cat ap)))))))
```

This FD can now be tested using the $test$ function:

```common-lisp
(test :item 't1)
```

# Questions on specific constitutents

The function $do-wh-constituents$ has a key $attrs$ optional parameter that points to the constituents on which we want the scope of the generated question.

```common-lisp
(do-wh-constituents 't39 :attrs '(partic lex-roles circum))
```
